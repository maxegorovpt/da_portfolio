# Introduction Simpson's Paradox

## 📌 Objective
To demonstratу Simpson’s Paradox where trends in aggregated data differ from those in subgroups and to investigate the presence of gender bias based on the most famous example university admission data by gender and department and other frequent examples

## 🧠 Background
Simpson’s Paradox occurs when a trend appears in several groups but reverses or disappears when the groups are combined. This problem is famously illustrated by the UC Berkeley admissions data, first widely discussed in Bickel et al. (1975):
[Simpson’s Paradox: UC Berkeley Admission Data](https://en.wikipedia.org/wiki/Simpson%27s_paradox#Berkeley_admissions_data)


## 🎲 Simulation

### 🎓 Case 1: UC Berkeley gender bias

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create a DataFrame with the given data on applicants and admission rates:

data = {
    'Department': ['A', 'B', 'C', 'D', 'E', 'F', 'Total'],
    'All': [933, 585, 918, 792, 584, 714, 4526],
    'Men': [825, 560, 325, 417, 191, 373, 2691],
    'Women': [108, 25, 593, 375, 393, 341, 1835],
    'All Admitted': ['64%', '63%', '35%', '34%', '25%', '6%', '39%'],
    'Men Admitted': ['62%', '63%', '37%', '33%', '28%', '6%', '45%'],
    'Women Admitted': ['82%', '68%', '34%', '35%', '24%', '7%', '30%']
}

df = pd.DataFrame(data)
df
```

```python
# Convert percentage strings to numeric values:
def convert_percent(value):
    return float(value.strip('%')) / 100

for col in ['All Admitted', 'Men Admitted', 'Women Admitted']:
    df[col] = df[col].apply(convert_percent)

# Calculate the number of admitted men and women per department:

df['Men Admitted Count'] = (df['Men'] * df['Men Admitted']).round().astype(int)
df['Women Admitted Count'] = (df['Women'] * df['Women Admitted']).round().astype(int)
df['Total Admitted Count'] = df['Men Admitted Count'] + df['Women Admitted Count']

df
```

```python
# Calculate overall admission rates by gender and compare:
total_men_admitted = df.loc[df['Department'] != 'Total', 'Men Admitted Count'].sum()
total_women_admitted = df.loc[df['Department'] != 'Total', 'Women Admitted Count'].sum()
total_men_applicants = df.loc[df['Department'] != 'Total', 'Men'].sum()
total_women_applicants = df.loc[df['Department'] != 'Total', 'Women'].sum()

overall_men_admission_rate = total_men_admitted / total_men_applicants
overall_women_admission_rate = total_women_admitted / total_women_applicants

# Print the results with explanatory text
print(f"Overall male admission rate: {overall_men_admission_rate:.2%}")
print(f"Overall female admission rate: {overall_women_admission_rate:.2%}")
```

```python
# Set visual style using Seaborn
sns.set_style("whitegrid")

plt.figure(figsize=(14, 7))
bar_width = 0.35
x = range(len(df))

# Define bar colors
men_color = '#4A90E2'      # Blue for men
women_color = '#D94F92'    # Pink for women

# Plot men bars (using existing 'Men Admitted' rate)
bars_men = plt.bar(
    [i - bar_width / 2 for i in x],
    df['Men Admitted'],
    width=bar_width,
    color=men_color,
    edgecolor='black',
    label='Men',
    alpha=0.9
)

# Plot women bars (using existing 'Women Admitted' rate)
bars_women = plt.bar(
    [i + bar_width / 2 for i in x],
    df['Women Admitted'],
    width=bar_width,
    color=women_color,
    edgecolor='black',
    label='Women',
    alpha=0.9
)

# Add text annotations with larger font size
for i in x:
    men_rate = df['Men Admitted'].iloc[i]
    women_rate = df['Women Admitted'].iloc[i]
    men_count = df['Men Admitted Count'].iloc[i]
    women_count = df['Women Admitted Count'].iloc[i]

    # Show % above each bar
    plt.text(i - bar_width/2, men_rate + 0.02,
             f"{men_rate:.0%}", ha='center', fontsize=14, weight='bold')
    plt.text(i + bar_width/2, women_rate + 0.02,
             f"{women_rate:.0%}", ha='center', fontsize=14, weight='bold')

    # Show admitted counts inside each bar
    plt.text(i - bar_width/2, men_rate / 2,
             f"{int(men_count)}", ha='center', va='center', fontsize=12, color='white', weight='bold')
    plt.text(i + bar_width/2, women_rate / 2,
             f"{int(women_count)}", ha='center', va='center', fontsize=12, color='white', weight='bold')

# Final formatting with bigger fonts
plt.xticks(x, df['Department'], fontsize=14)
plt.yticks(fontsize=13)
plt.ylabel("Admission Rate", fontsize=16)
plt.ylim(0, 1.1)
plt.title("Admission Rates by Department and Gender", fontsize=18, weight='bold', pad=20)
plt.legend(title="Gender", fontsize=14, title_fontsize=16)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Set seaborn style and font that supports emoji
sns.set_style("whitegrid")
plt.rcParams['font.family'] = 'DejaVu Sans'

# Admission rates data
labels = ['Men', 'Women']
rates = [overall_men_admission_rate, overall_women_admission_rate]
colors = ['steelblue', 'orchid']

# Total admitted counts and applicants (replace with your actual values)
total_men_admitted = df.loc[df['Department'] != 'Total', 'Men Admitted Count'].sum()
total_women_admitted = df.loc[df['Department'] != 'Total', 'Women Admitted Count'].sum()
total_men_applicants = df.loc[df['Department'] != 'Total', 'Men'].sum()
total_women_applicants = df.loc[df['Department'] != 'Total', 'Women'].sum()

# Create the figure
plt.figure(figsize=(8, 6))
bars = plt.bar(labels, rates, color=colors, width=0.5)

# Annotate each bar with the exact percentage, admitted count and total applicants
for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.01,
             f'{height:.1%}', ha='center', va='bottom',
             fontsize=14, fontweight='bold', color='black')

    # Add admitted counts inside the bar (white text)
    admitted_count = total_men_admitted if i == 0 else total_women_admitted
    plt.text(bar.get_x() + bar.get_width() / 2, height / 2,
             f'{int(admitted_count)} admitted', ha='center', va='center',
             fontsize=12, fontweight='bold', color='white')

    # Add total applicants count below the bar
    applicants_count = total_men_applicants if i == 0 else total_women_applicants
    plt.text(bar.get_x() + bar.get_width() / 2, -0.05,
             f'{int(applicants_count)} applicants', ha='center', va='top',
             fontsize=12, fontweight='bold', color='gray')

# Title and labels
plt.title("Overall Admission Rates by Gender", fontsize=18, fontweight='bold')
plt.ylabel("Admission Rate", fontsize=14)
plt.ylim(-0.1, 0.6)  # extend y-limit down to fit applicants count
plt.xticks(fontsize=14)
plt.yticks(fontsize=13)

# Add explanatory annotation
plt.text(0.5, 0.53,
         "Although women had higher admission rates in most departments,\n"
         "they applied more to departments with lower overall admission rates.",
         ha='center', fontsize=12,
         bbox=dict(facecolor='lightyellow', edgecolor='gray', boxstyle='round,pad=0.4'))

# Add subtle grid
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### 🏥 Case 2: Medical Treatment

Imagine you need to choose which hospital to go to, and you only have high‑level survival data for each one. You see that Hospital A has 900 survivors out of 1,000 patients (90% survival), while Hospital B has 800 survivors out of 1,000 patients (80% survival). At first glance, Hospital A looks clearly better.

```python
# Data
hospitals = ["A", "B"]
rates = [0.90, 0.80]  # 90% and 80%
survived = [900, 800]
total_patients = [1000, 1000]

# Use a light, minimal style
plt.style.use("default")  # or "seaborn-v0_8-deep" if you like it a bit softer

fig, axs = plt.subplots(2, 1, figsize=(8, 6), gridspec_kw={"height_ratios": [3, 1]})

# 1) Light bar chart (top)
ax1 = axs[0]

colors = ["#6CA6CD", "#FFA07A"]  # soft blue and light salmon

bars = ax1.bar(
    hospitals, rates,
    color=colors,
    width=0.5,
    edgecolor="none",      # remove strong outline
    linewidth=0.5          # very thin if needed
)

ax1.set_title(
    "Overall survival by hospital – Hospital A appears better",
    fontsize=13,
    pad=10
)
ax1.set_ylabel("Survival rate", fontsize=11)
ax1.set_ylim(0, 1.05)

# Very light grid, no bold spines
ax1.grid(axis="y", alpha=0.15, linestyle="-", linewidth=0.5)
ax1.spines["top"].set_visible(False)
ax1.spines["right"].set_visible(False)
ax1.spines["left"].set_linewidth(0.5)
ax1.spines["bottom"].set_linewidth(0.5)

# Lighter labels and tick marks
ax1.tick_params(axis="both", which="major", labelsize=10, width=0.5)

# Annotate percentages lightly
for i, r in enumerate(rates):
    ax1.text(
        i, r + 0.03,
        f"{r:.1%}",
        ha="center",
        fontsize=11,
        color="gray",
        alpha=0.9
    )


# 2) Table with numbers (bottom)
ax2 = axs[1]
ax2.axis("tight")
ax2.axis("off")

df = pd.DataFrame({
    "Hospital": ["A", "B"],
    "Total patients": total_patients,
    "Survived": survived,
    "Survival rate": [f"{r:.1%}" for r in rates]
})

cell_text = df.values.tolist()
columns = df.columns.tolist()

table = ax2.table(
    cellText=cell_text,
    colLabels=columns,
    cellLoc="center",
    loc="center",
    bbox=[0.1, 0.1, 0.8, 0.8]
)

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.5)

# Light header, no bold
for j in range(len(columns)):
    table[(0, j)].set_facecolor("#f7f7f7")
    table[(0, j)].set_text_props(weight="normal", fontsize=10)

plt.tight_layout()
plt.show()
```

But when you dig deeper and break the data down by patients’ health status—good or poor—you suddenly see the opposite pattern in every subgroup. In poor‑health patients, Hospital A has 30 survivors out of 100 (30%), while Hospital B has 210 survivors out of 400 (52.5%), so B performs much better. In good‑health patients, Hospital A has 870 survivors out of 900 (about 96.7%), while Hospital B has 590 survivors out of 600 (about 98.3%), so B is still slightly better or at least not worse.  

```python
df_breakdown = pd.DataFrame({
    "Hospital": ["A", "A", "B", "B"],
    "Health_status": ["Poor", "Good", "Poor", "Good"],
    "Patients": [100, 900, 400, 600],
    "Survived": [30, 870, 210, 590],
})
df_breakdown["Survival_rate"] = df_breakdown["Survived"] / df_breakdown["Patients"]

# 2.2) Pivot for charting
piv = df_breakdown.pivot(
    index="Health_status",
    columns="Hospital",
    values="Survival_rate"
)

# 2.3) By‑health bar chart + table
fig2, axs2 = plt.subplots(2, 1, figsize=(9, 7), gridspec_kw={"height_ratios": [3, 1]})

ax3 = axs2[0]
x = [0, 1]  # Poor, Good
width = 0.3

ax3.bar(
    [xi - width/2 for xi in x],
    piv["A"],
    width=width,
    label="Hospital A",
    color="#6CA6CD",
    edgecolor="none",
    alpha=0.9
)
ax3.bar(
    [xi + width/2 for xi in x],
    piv["B"],
    width=width,
    label="Hospital B",
    color="#FFA07A",
    edgecolor="none",
    alpha=0.9
)

ax3.set_title(
    "Survival by health status – B better in each group",
    fontsize=13, pad=12
)
ax3.set_ylabel("Survival rate", fontsize=11)
ax3.set_ylim(0, 1.05)
ax3.set_xticks(x)
ax3.set_xticklabels(["Poor health", "Good health"])
ax3.tick_params(axis="both", which="major", labelsize=10)
ax3.grid(axis="y", alpha=0.15, linestyle="-", linewidth=0.5)
ax3.spines["top"].set_visible(False)
ax3.spines["right"].set_visible(False)
ax3.legend(fontsize=10)

# Annotate rates
for i, health in enumerate(piv.index):
    for j, hosp in enumerate(["A", "B"]):
        r = piv.loc[health, hosp]
        ax3.text(
            i + (-1 if j==0 else 1)*0.15,
            r + 0.03,
            f"{r:.1%}",
            ha="center",
            fontsize=9,
            color="gray",
            alpha=0.9
        )

# 2.4) Table by health status
ax4 = axs2[1]
ax4.axis("tight")
ax4.axis("off")

# Format for table
df_table = df_breakdown.copy()
df_table["Survival_rate_str"] = df_table["Survival_rate"].apply(lambda x: f"{x:.1%}")
df_table = df_table.drop(columns=["Survival_rate"])

cell_text_break = df_table.values.tolist()
columns_break = df_table.columns.tolist()

table_break = ax4.table(
    cellText=cell_text_break,
    colLabels=columns_break,
    cellLoc="center",
    loc="center",
    bbox=[0.1, 0.1, 0.8, 0.8]
)

table_break.auto_set_font_size(False)
table_break.set_fontsize(9)
table_break.scale(1.2, 1.8)
for j in range(len(columns_break)):
    table_break[(0, j)].set_facecolor("#f7f7f7")

plt.tight_layout()
plt.show()
```

## ☝️Outcomes
Simpson’s Paradox shows how aggregated data can hide or reverse true patterns. It highlights the need to think causally, understand how the data is generated, and decide whether to keep groups separate or combine them in order to draw accurate conclusions.

## 🤔 What to keep in mind in real life:
- Don’t judge decisions by aggregate data alone. Overall numbers can hide important details that affect fairness and accuracy.
- Look deeper into subgroups or categories. Context matters — different groups may have different behaviors or conditions influencing the results.
- Be cautious in interpreting “bias” or “discrimination.” Sometimes, apparent disparities arise from underlying factors like preferences or distribution across groups.
- Make data-driven decisions thoughtfully. Dig into the data structure before drawing conclusions or making policy changes.
- Communicate clearly. When presenting data, highlight how group-level effects might differ from overall trends to avoid misunderstandings.

**In short, real-world data is often complex, and understanding these nuances helps create fairer, more effective analyses and decisions.**
