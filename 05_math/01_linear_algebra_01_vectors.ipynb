# Vectors

## How to set up a vector

A vector is an ordered set of numeric data.

Python has a library designed, among other things, for working with vectors. It's called Numeric Python, or NumPy for short. It implements a huge number of functions for mathematical calculations.

A vector in NumPy is an object of a special type called an array. To create such an object, you call the NumPy library function of the same name:

```python
import numpy as np
```

```python
v = np.array([2, 15, 6])
```

## Addition and subtraction of vectors

In arithmetic, we add and subtract numbers, in algebra, we add and subtract variables, and in linear algebra, we perform the same operations with vectors. Moreover, vectors can be added and subtracted, as well as defined, algebraically and geometrically.

```python
# addition
january = np.array([648, 371, 338.4, 512, 0, 437, 11])
february = np.array([112, 380, 516, 294, 108, 380, 65])
march = np.array([506, 0, 412, 98, 304, 811, 56])

s = january + february + march
print(s)
```

```python
# substraction
income = np.array([2.8, 4.31, 3.75, 5.4, 5.98, 4.9, 3.11, 4.5, 5.24, 4, 3.8, 6.11])
outcome = np.array([2.3, 2.3, 2.2, 4.2, 3.7, 3.25, 3.2, 4.11, 3.87, 2.9, 2.65, 3.3])
print(income - outcome)
```

```python
# multiplication
a = np.array([3.7, 4.8, 4.2, 7.2, 7.6])
c = 1.2 * a
print(c)
```

```python
# division
salary = np.array([60000, 100000, 80000, 65000, 73000])
print(salary / 15)
```

## Dot product of vectors

The scalar product of two vectors is the product of their lengths and the cosine of the angle between them.

```python
a = np.array([5, 4, 8])
b = np.array([7, 12, 6])
c = a @ b
print(c)
```

## Linear combination of vectors

Vectors, like other objects, form sets. One way to define a set is to define properties that all its elements satisfy. These properties must uniquely identify the elements of the set: if elements possess the set's properties, they are included in it; if they lack at least one, they are not included.

```python
a = np.array([45, 21, 6, 0, 18, 5])
b = np.array([-6, 11, 8, 32, 9, 115])
c = np.array([1, 2, 3, 4, 5, 6])
d = np.array([318, 4, 0, 0, 0, -7])

k1, k2, k3, k4 = 3, 11, -5, 7

l = k1 * a + k2 * b + k3 * c + k4 * d
print(l)
```

## How to calculate vector's norm

```python
# L1 норма
import numpy as np

k = np.array([25, -41, -31, 66])
l1 = np.linalg.norm(k, 1) # Вызываем функцию norm() из np.linalg .
print(l1)
```

```python
# L1 норма
from numpy import array # Подключение функции array() из numpy.
from numpy.linalg import norm # Подключение функции norm() из numpy.linalg.

k = array([8, -16, -4, 5]) # Теперь не нужно писать np.
l1 = norm(k, 1)
print(l1)
```

```python
# L2 норма
from numpy import array
from numpy.linalg import norm 

k = array([8, -16, -4, 5])
l2 = norm(k, 2)
print(l2)
```

```python
# Максимальная норма
from numpy import array
from numpy import inf  # Подключаем бесконечность.
from numpy.linalg import norm

k = array([8, -16, -4, 5])
maxnorm = norm(k, inf)
print(maxnorm)
```

## How to find the angle between vectors 

```python
from numpy import array
from numpy import arccos
from numpy import rad2deg  
from numpy.linalg import norm

a = array([4, 8, 17])
b = array([-2, 3, 7])
cos_angle = a @ b / norm(a) / norm(b)
angle = arccos(cos_angle) 
print("Косинус угла между a и b:", cos_angle) # В кавычках, через запятую.
print("Сам угол:", rad2deg(angle))  # В кавычках, через запятую.
```

## Calculating Distances in Python

```python
from numpy import array
from numpy import inf
from numpy.linalg import norm

a = array([22, 4, -6, 3.9, 7])
b = array([12, -0.5, 6, 42, 1])

r_1 = norm(a - b, 1)  # Вычисление L1 расстояния.
r_2 = norm(a - b, 2)  # Вычисление L2 расстояния.
r_max = norm(a - b, inf)  # Вычисление максимального расстояния.

print("L1 расстояние между векторами a и b:", r_1)  # Вывод на экран L1 расстояния.
print("L2 расстояние между векторами a и b:", r_2)  # Вывод на экран L2 расстояния.
print(
    "Максимальное расстояние между векторами a и b:", r_max
)  # Вывод на экран максимального расстояния.
```

## Cosine distance

```python
from numpy import array
from numpy.linalg import norm

o_1 = array([-4, 6, 12])
o_2 = array([2, -8, 3])
o_3 = array([5, 3, -1])
cos_1_2 = o_1 @ o_2 / norm(o_1) / norm(o_2)
cos_1_3 = o_1 @ o_3 / norm(o_1) / norm(o_3)
cos_2_3 = o_2 @ o_3 / norm(o_2) / norm(o_3)
print("Отзывы 1 и 2:", cos_1_2)
print("Отзывы 1 и 3:", cos_1_3)
print("Отзывы 2 и 3:", cos_2_3)
```
